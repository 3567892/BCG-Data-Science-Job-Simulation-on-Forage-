```python
#Changing date columns from object to date
date_columns_client = ['date_activ','date_end','date_modif_prod','date_renewal']

client_df[date_columns_client] = client_df[date_columns_client].apply(pd.to_datetime, format= '%m/%d/%y')

#Changing has_gas columns from object to int64
has_gas_columns = ['has_gas']
client_df[has_gas_columns] = client_df[has_gas_columns].apply(pd.to_numeric, errors = 'coerce' , downcast = 'integer')
client_df.info()
```

```python
#Changing date columns from object to date

date_columns_price = ['price_date']
price_df[date_columns_price] = price_df[date_columns_price].apply(pd.to_datetime,format = '%m/%d/%y')
price_df.info()
```

```python
#Creating an extra for contraction duration
client_df['contract_duration'] = (client_df['date_end'] - client_df['date_activ']).dt.days
client_df.head(5)
```

Data Visualisations

```python
#random samples data due to dataset being large
n_samples = 50

random_sample = client_df.sample(n=n_samples, random_state=42)

#function to plot columns 
def plot_consumption_analysis(random_sample, columns, target='churn', bins=5, y_bins=100):
    num_columns = len(columns)
    rows = (num_columns + 1) // 2

    plt.figure(figsize=(20, 10 * rows))
#creates a histogramplot for numeric columns and countplot for categoric variables
    for i, column in enumerate(columns):
        plt.subplot(rows, 2, i + 1)

        if random_sample[column].dtype == 'float64' or random_sample[column].dtype == 'int64':
            sns.histplot(data=random_sample, x=column, hue=target, bins=bins)
            plt.title(f'{column} Distribution')
            plt.ticklabel_format(style='plain', axis='x')

        else:
            sns.countplot(data=random_sample, x=column, hue=target)
            plt.title(f'{column} Count')
 #rotates my x-axes
        plt.xticks(rotation=90)

    plt.tight_layout()
    plt.show()

```

```python
columns_to_plot = ['cons_12m', 'cons_gas_12m', 'cons_last_month', 'forecast_cons_12m', 'forecast_cons_year','num_years_antig', 'contract_duration']
plot_consumption_analysis(random_sample,columns=columns_to_plot)
```

```python
columns_to_plot = [ 'channel_sales']
plot_consumption_analysis(random_sample,columns=columns_to_plot)
```

```python
#Joining both csvs

merged_data = pd.merge(client_df,price_df, on ='id', how = 'inner')
merged_data.head()
```

```python

#Plot to view if there are ny patterns in the pricing for diferent churns
sns.scatterplot(x='price_off_peak_var', y='price_peak_var', hue='churn', data=merged_data)
plt.title('Price vs. Churn')
plt.show()
```

```python
#PLot to view the electricity consumption for both churned and non non-churned
sns.boxplot(x='churn', y='cons_12m', data=merged_data)
plt.title('Electricity Consumption vs. Churn')
plt.show()
```

```python
#Plot to display code of the electricity campaign the customer first subscribed to for both churned and non-churned

sns.countplot(x='origin_up', hue='churn', data=merged_data)
plt.title('Origin vs. Churn')
plt.xticks(rotation=90)
plt.show()
```

```python
#Created Time Plot to view how electricity consumption changed over time
sns.lineplot(x='date_activ', y='cons_12m', hue='churn', data=merged_data)
plt.title('Electricity Consumption Over Time vs. Churn')
plt.show()
```
