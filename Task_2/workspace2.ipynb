Feature Engineering

```python
import pandas as pd
```

```python
clean_df = pd.read_csv('clean_data_after_eda.csv')
clean_df.head()
```

```python
clean_df.info()
```

```python
#Changing date columns from object to date
date_columns_client = ['date_activ','date_end','date_modif_prod','date_renewal']

clean_df[date_columns_client] = clean_df[date_columns_client].apply(pd.to_datetime, format= '%m/%d/%y')
```

```python
#CHecking for unique values
clean_df.nunique()
```

```python
#Creating new month columns 
clean_df['date_activ_month'] = clean_df['date_activ'].dt.month

clean_df['date_end_month'] = clean_df['date_end'].dt.month

clean_df['date_modif_prod_month'] = clean_df['date_modif_prod'].dt.month

clean_df['date_renewal_month'] = clean_df['date_renewal'].dt.month

#Creating year columns

clean_df['date_activ_year'] = clean_df['date_activ'].dt.year

clean_df['date_end_year'] = clean_df['date_end'].dt.year

clean_df['date_modif_prod_year'] = clean_df['date_modif_prod'].dt.year

clean_df['date_renewal_year'] = clean_df['date_renewal'].dt.year

#Changing has_gas columns from object to int64
#lean_df['has_gas'] = clean_df['has_gas'].astype(int)

```

```python
#Creating an extra for contraction duration
clean_df['contract_duration'] = (clean_df['date_end'] - clean_df['date_activ']).dt.days
clean_df.head(5)
```

```python
#Column for difference between prices for off peak and peak (energy)
clean_df['var_year_price_op_difference_var'] = (clean_df['var_year_price_off_peak_var'] - clean_df['var_year_price_peak_var'])

#Column for difference between prices for off peak and mid - peak(enrgy)
clean_df['var_year_price_om_difference_var'] = (clean_df['var_year_price_off_peak_var'] - clean_df['var_year_price_mid_peak_var'])

#Column for difference between prices for  peak and mid peak(energy)

clean_df['var_year_price_pm_difference_var'] = (clean_df['var_year_price_peak_var'] - clean_df['var_year_price_mid_peak'])

#Column for difference between prices for off peak and peak(power)
clean_df['var_year_price_op_difference_var'] = (clean_df['var_year_price_off_peak_fix'] - clean_df['var_year_price_peak_fix'])

#Column for difference between prices for off peak and mid - peak(power)
clean_df['var_year_price_om_difference_var'] = (clean_df['var_year_price_off_peak_fix'] - clean_df['var_year_price_mid_peak_fix'])

#Column for difference between prices for  peak and mid peak(power)

clean_df['var_year_price_pm_difference_var'] = (clean_df['var_year_price_peak_fix'] - clean_df['var_year_price_mid_peak_fix'])


```

```python
#Dropping the these columns since they do not seem to contribute to why customers could be churning


clean_df = clean_df.drop('margin_net_pow_ele', axis=1)


clean_df = clean_df.drop('margin_gross_pow_ele', axis=1)
```
